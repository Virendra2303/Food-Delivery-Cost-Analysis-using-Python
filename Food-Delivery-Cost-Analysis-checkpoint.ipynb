#### The dataset contains 1,000 entries and 12 columns, with missing values in Discounts and Offers columns.<br> Now, we need to perform some data cleaning and preparation

```python
df["Discounts and Offers"].fillna("None",inplace=True)
```

```python
df.info()
```

### Below are the necessary cleaning steps we need to take:
<br>1)
Convert “Order Date and Time” and “Delivery Date and Time” to a datetime format.<br>
2)
Convert “Discounts and Offers” to a consistent numeric value (if applicable) or calculate the discount amounts.


```python
from datetime import datetime
df["Order Date and Time"]=pd.to_datetime(df["Order Date and Time"])
df["Delivery Date and Time"]=pd.to_datetime(df["Delivery Date and Time"])
def obtainvalues(str):
    if "off" in str:
        return float(str.split(" ")[0])
    elif "%" in str:
        return float(str.split("%")[0])
    else:
        return 0.0
df["Discount Percentage"]=df["Discounts and Offers"].apply(obtainvalues)
df["Discount Amount"]=df.apply(lambda x: (x["Order Value"]*x["Discount Percentage"]/100)
                              if x["Discount Percentage"]>1
                              else x["Discount Percentage"],axis=1)
print(df[['Order Value', 'Discounts and Offers', 'Discount Percentage', 'Discount Amount']].head(),df.dtypes)
```

### Costs associated with each order:
- **Delivery Fee**: The fee charged for delivering the order.
- **Payment Processing Fee**: The fee for processing the payment.
- **Discount Amount**: The discount provided on the order.der.

```python
df["Cost"]=df["Delivery Fee"]+df["Payment Processing Fee"]+df["Discount Amount"]
df["Revenue"]=df["Commission Fee"]
df["Profit"]=df["Revenue"]-df["Cost"]
total_cost=df["Cost"].sum()
total_revenue=df["Revenue"].sum()
total_profit=df["Profit"].sum()
total_order=df["Order ID"].count()

sales_value={"Total Order":total_order,
            "Total Cost":total_cost,
            "Total Revenue":total_revenue,
            "Total Profit":total_profit}
print(sales_value)
```

### Based on the analysis, here are the overall metrics for the food delivery operations:

- **Total Orders**: 1,000
- **Total Costs**: 232,709.85 INR (including delivery fees, payment processing fees, and discounts)
- **Total Revenue** (from Commission Fees): 126,990 INR
- **Total Profit**: -105,719.85 INR

#### let’s compare total revenue, total costs, and total profit

```python
import matplotlib.pyplot as plt
import seaborn as sns
totals = ['Total Revenue', 'Total Costs', 'Total Profit']
values = [total_revenue, total_cost, total_profit]
plt.figure(figsize=(10,6))
plt.bar(x=totals,height=values,color=["green","red","yellow"],label=totals)
plt.legend()
plt.ylabel("Amount(Rs)")
plt.title("Total Cost,Revenue And Profit")
plt.show()
```

##### let’s have a look at the proportion of total costs:

```python
costs = df[['Delivery Fee', 'Payment Processing Fee', 'Discount Amount']].sum()
plt.figure(figsize=(10,6))
plt.pie(x=costs.values,autopct="%0.2f",labels=costs.index)
plt.title("Proportion of Total Costs in Food Delivery")
plt.show()
```
